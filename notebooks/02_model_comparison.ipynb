# Model Comparison: PyTorch vs Scikit-Learn

This notebook compares our deep learning BiLSTM against the TF-IDF baseline models for both **Binary** and **Multi-Class** tasks.

```python
import json
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

ROOT = Path.cwd().parent
OUTPUTS = ROOT / "outputs"
```

## 1. Binary Classification Comparison

```python
binary_models = {
    "PyTorch BiLSTM": OUTPUTS / "binary",
    "Logistic Regression": OUTPUTS / "logistic" / "binary",
    "Random Forest": OUTPUTS / "rf" / "binary"
}

binary_metrics = {}

for name, path in binary_models.items():
    metrics_file = path / "metrics.json"
    if metrics_file.exists():
        with open(metrics_file, "r") as f:
            data = json.load(f)
            binary_metrics[name] = {
                "Accuracy": data.get("accuracy", 0),
                "Precision": data.get("precision", 0),
                "Recall": data.get("recall", 0),
                "F1 Score": data.get("f1", 0),
                "ROC AUC": data.get("roc_auc", 0)
            }

if binary_metrics:
    df_binary = pd.DataFrame(binary_metrics).T
    display(df_binary.style.highlight_max(color="lightgreen", axis=0))
else:
    print("Binary metrics not found. Please run the training scripts for binary classification.")
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle("Binary Confusion Matrices", fontsize=16)

for ax, (name, path) in zip(axes, binary_models.items()):
    img_path = path / "confusion_matrix.png"
    if img_path.exists():
        img = mpimg.imread(img_path)
        ax.imshow(img)
        ax.axis("off")
        ax.set_title(name)
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle("Binary ROC Curves", fontsize=16)

for ax, (name, path) in zip(axes, binary_models.items()):
    img_path = path / "roc_curve.png"
    if img_path.exists():
        img = mpimg.imread(img_path)
        ax.imshow(img)
        ax.axis("off")
        ax.set_title(name)
plt.tight_layout()
plt.show()
```

## 2. Multi-Class Classification Comparison

```python
multi_models = {
    "PyTorch BiLSTM": OUTPUTS / "multiclass",
    "Logistic Regression": OUTPUTS / "logistic" / "multiclass",
    "Random Forest": OUTPUTS / "rf" / "multiclass"
}

multi_metrics = {}

for name, path in multi_models.items():
    metrics_file = path / "metrics.json"
    if metrics_file.exists():
        with open(metrics_file, "r") as f:
            data = json.load(f)
            multi_metrics[name] = {
                "Accuracy": data.get("accuracy", 0),
                "Macro Precision": data.get("classification_report", {}).get("macro avg", {}).get("precision", 0),
                "Macro Recall": data.get("classification_report", {}).get("macro avg", {}).get("recall", 0),
                "Macro F1": data.get("macro_f1", 0),
                "Weighted F1": data.get("weighted_f1", 0),
            }

if multi_metrics:
    df_multi = pd.DataFrame(multi_metrics).T
    display(df_multi.style.highlight_max(color="lightgreen", axis=0))
else:
    print("Multi-class metrics not found. Please run the training scripts for multiclass classification.")
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle("Multi-class Confusion Matrices", fontsize=16)

for ax, (name, path) in zip(axes, multi_models.items()):
    img_path = path / "confusion_matrix.png"
    if img_path.exists():
        img = mpimg.imread(img_path)
        ax.imshow(img)
        ax.axis("off")
        ax.set_title(name)
plt.tight_layout()
plt.show()
```
